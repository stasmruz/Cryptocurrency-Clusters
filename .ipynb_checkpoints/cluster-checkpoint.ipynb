```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#import csv file
crypto= pd.read_csv('crypto_data.csv')
crypto.head (-5)
```

```python
#find when istrading =true and drop Nas
crypto= crypto.loc[crypto['IsTrading']==True]

crypto=crypto.drop('IsTrading',axis=1)
crypto.head(-5)
```

```python
crypto= crypto.dropna()
crypto.shape
```

```python
crypto=crypto[crypto['TotalCoinsMined']>0]
crypto.head(-5)
```

```python
crypto = crypto.drop('Unnamed: 0', 1)
crypto.head(-5)
```

```python
# get dummies
crypto = crypto.drop('CoinName', axis=1)
x=pd.get_dummies(data=crypto, columns=['Algorithm','ProofType'])
x.head(-1)
```

```python
#set up scale use Pca
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
PC = PCA(n_components=.90)
p_c = PC.fit_transform(x_scaled)
```

```python
#use TSNE test shape should reduce it
tsne_ = TSNE(perplexity=50)
tsne_ft = tsne_.fit_transform(p_c)
tsne_ft.shape
```

```python
# create a scatter plot to see if there are distant clusters
x = tsne_ft[:,0]
y = tsne_ft[:,1]

plt.scatter(x, y)
plt.show()
```

```python
#create df for tsne for model
tsne_df = pd.DataFrame(data=tsne_ft)
#create a model to see how much clusters is best option
model = KMeans(n_clusters=5, random_state=5)
model.fit(tsne_df)
predictions = model.predict(tsne_df)
# plot 
plt.scatter(x=tsne_df[0], y=tsne_df[1],c=predictions)
plt.show()
```

```python
#set up floop to deterimte intertia for each K,  then plot elbow plot 
I = []
k = list(range(1,10))

for a in k:
    kmeans = KMeans(n_clusters=a, random_state=0)
    kmeans.fit(p_c)
    I.append(kmeans.inertia_)

elbow = pd.DataFrame({'k': k, 'I': I})
elbow.plot.line(x='k', y='I')
plt.ylabel('Inertia')
plt.title('Elbow Plot')
plt.show()
```

Although there is an outlier, I do beleive cryptocurrencies can be clustered together. The best fit that I have tested; 5 clusters shows best results.
